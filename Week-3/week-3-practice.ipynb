## Natural language processing (NLP)

Natural Language Processing (NLP) is a branch of artificial intelligence that enables computers to understand, interpret, and generate human language.

```python
# import necessary libraries

import nltk

nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("averaged_perceptron_tagger")
nltk.download("averaged_perceptron_tagger_eng")
```

```python
# Practice - 1: Clean and Preprocess Sample Text

import re

text = "Hello!!! My name is Ridesha. I love AI and Python :)"

print("Original Text:")
print(text)

clean_text = text.lower()
clean_text = re.sub(r"[^a-zA-Z\s]", "", clean_text)

print("\nCleaned Text:")
print(clean_text) 
```

```python
# Practice 2: Tokenization

from nltk.tokenize import word_tokenize

text = "I love learning Natural Language Processing."
tokens = word_tokenize(text)

print(tokens)
```

```python
# Practice 3: Stop Word Removal

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "I am learning AI in Nepal"

tokens = word_tokenize(text)

stop_words = set(stopwords.words("english"))

filtered_words = [
    word for word in tokens
    if word.lower() not in stop_words
]

print(filtered_words)
```

```python
# Practice 4: Stemming

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["playing", "played", "plays", "studies"]

for word in words:
    print(word, "->", stemmer.stem(word))
```

```python
# Practice 5: Lemmatization

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["playing", "played", "studies"]

for word in words:
    print(word, "->", lemmatizer.lemmatize(word))
```

```python
# Practice 6: Part-of-Speech Tagging

from nltk import pos_tag
from nltk.tokenize import word_tokenize

text = "Ridesha writes Python code"

tokens = word_tokenize(text)

print(pos_tag(tokens))
```

```python
# Practice 7: Simple Sentiment Analysis

from textblob import TextBlob

text = "This internship is good and I am learning a lot."

blob = TextBlob(text)

print("Sentiment Score:", blob.sentiment.polarity)
```
