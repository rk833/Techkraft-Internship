```python
# Load the customer reviews dataset

import pandas as pd

df = pd.read_csv("customer_reviews.csv")

df.head()
```

```python
# Text Cleaning 

import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    return text

df["clean_review"] = df["review"].apply(clean_text)

df[["review", "clean_review"]]
```

#### Observation

The text cleaning step converted all text to lowercase and removed punctuation and special characters. This made the text more consistent and easier to process. Cleaning text helps reduce noise and improves the quality of later NLP tasks such as tokenization and sentiment analysis.


```python
# Tokenization

from nltk.tokenize import word_tokenize

df["tokens"] = df["clean_review"].apply(word_tokenize)

df[["clean_review", "tokens"]]
```

#### Observation

The tokenization process split each review into individual words called tokens. This allows the text to be analyzed at the word level and prepares it for further processing such as stop word removal and lemmatization.


```python
# Remove Stopwords

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

def remove_stopwords(tokens):
    return [
        word for word in tokens
        if word not in stop_words
    ]

df["filtered_tokens"] = df["tokens"].apply(remove_stopwords)

df[["tokens", "filtered_tokens"]]
```

####
 Observation

Common words such as "is", "and", and "the" were removed from the reviews. These words occur frequently but often carry little meaning. Removing stop words helps focus on the most informative words in the text.


```python
# Lemmatization

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_words(tokens):
    return [
        lemmatizer.lemmatize(word)
        for word in tokens
    ]

df["lemmatized"] = df["filtered_tokens"].apply(lemmatize_words)

df[["filtered_tokens", "lemmatized"]]
```

#### Observation

Lemmatization converted words into their base dictionary form. This helps group similar words together and reduces redundancy in the text. Compared to stemming, lemmatization produces more meaningful and readable words.

```python
# Sentiment Analysis

from textblob import TextBlob

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df["sentiment_score"] = df["review"].apply(get_sentiment)

df[["review", "sentiment_score"]]
```

#### Observation

Sentiment analysis assigned a polarity score to each review. Positive reviews received positive scores, while negative reviews received negative scores. This demonstrates how NLP can be used to automatically understand customer opinions and feedback.

```python
# Sentiment label

def sentiment_label(score):
    if score > 0:
        return "Positive"
    elif score < 0:
        return "Negative"
    else:
        return "Neutral"

df["sentiment"] = df["sentiment_score"].apply(sentiment_label)

df[["review", "sentiment"]]
```

#### Observation

The sentiment scores were converted into labels: Positive, Negative, and Neutral. This makes the results easier to interpret and provides a simple way to categorize customer reviews.

```python
# Summary of Sentiment Analysis

print("Total Reviews:", len(df))

print("\nSentiment Counts:")
print(df["sentiment"].value_counts())
```

```python
# Save results to a new CSV file

df.to_csv("processed_reviews.csv", index=False)

print("Processed file saved successfully")
```
