```python
# Write small programs using lists and dictionaries

# This program demonstrates basic operations on a list of numbers.
numbers = [10, 20, 30, 40, 50]

print("List:", numbers)
print("Sum:", sum(numbers))
print("Largest:", max(numbers))
print("Smallest:", min(numbers))
```

```python
# This program demonstrates basic operations using tuples.

student = ("Ridesha", 22, "AI Intern")

print("Name:", student[0])
print("Age:", student[1])
print("Role:", student[2])
```

```python
# This program demonstrates basic operations using dictionaries.

student = {
    "name": "Ridesha",
    "age": 22,
    "marks": 90
}

for key, value in student.items():
    print(key, ":", value)
```

```python
# This program demonstrates basic set operations.

python_students = {"Ridesha", "Asha", "Sita"}
ai_students = {"Ram", "Asha", "Ridesha"}

print("Union:", python_students | ai_students)
print("Common:", python_students & ai_students)
```

```python
# Solve simple logic problems

number = int(input("Enter a number: "))

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Convert data to json

import csv
import json

students = []

with open("students.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        students.append(row)

with open("students.json", "w") as file:
    json.dump(students, file, indent=4)

print("JSON file created successfully")
```
